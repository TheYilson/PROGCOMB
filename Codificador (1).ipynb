```python
def transformar_palabra(palabra):
    voc = "aeiou"
    conso = "bcdfghjklmnpqrstvwxyz"

    nueva_palabra = ""

    for letra in palabra:
        if letra in voc:
            indice = voc.find(letra)
            nueva_letra = voc[indice-1] if indice-1 >= 0 else voc[-1]
        elif letra in conso:
            indice = conso.find(letra)
            nueva_letra = conso[indice+1] if indice+1 < len(conso) else conso[0]
        else:
            nueva_letra = letra

        nueva_palabra += nueva_letra

    return nueva_palabra

palabra = input("Ingrese una palabra: ")
print((transformar_palabra(palabra)))
```

```python
import pandas as pd
import random

nom=["Andres","Daniela","Carlos","Naiver","Santi","Leidy","Daniel","Yilson","Felipe","Manuel"]
ape=["torre","Maestre","Efren","Peres","Blachar","Gutierres","Moscote","Carrillo","Bolaños","Brito"]
pesos12=[12,17,18,20,24,27,34,37,65,45]
alturas12=[1.43,1.57,1.52,1.67,1.70,1.73,1.75,1.84,1.87,1.90]

user = {"Nombre:":[], "Apellido:":[], "Peso:":[], "Altura:":[], "IMC:":[], "Descripciones:":[]}

def generate_data():
  Nombre = random.choice(nom)
  Apellido = random.choice(ape)
  Peso = random.choice(pesos12)
  Altura = random.choice(alturas12)
  return Nombre, Apellido, Peso, Altura

def calculate_imc(peso, altura):
  imc = int(peso/(altura)**2)
  return imc

def describe_imc(imc):
  if imc<16.9:
    valor="Muy bajo peso"
  elif imc>=17.0 and imc<=18.4:
    valor="Bajo peso"
  elif imc>=18.5 and imc<=24.9:
    valor="Normal"
  elif imc>=25.0 and imc<=29.9:
    valor="Sobrepeso"
  elif imc>=30.0 and imc<=34.9:
    valor="Obesidad"
  elif imc>=35.0 and imc<=39.9:
    valor="Obesidad marcada"
  elif imc>=40.0:
    valor="Obesidad mórbida"
  return valor

for i in range(10):
  Nombre, Apellido, Peso, Altura = generate_data()
  IMC = calculate_imc(Peso, Altura)
  Descripcion = describe_imc(IMC)

  user["Nombre:"].append(Nombre)
  user["Apellido:"].append(Apellido)
  user["Peso:"].append(Peso)
  user["Altura:"].append(Altura)
  user["IMC:"].append(IMC)
  user["Descripciones:"].append(Descripcion)
pd.DataFrame.from_dict(user)
```
